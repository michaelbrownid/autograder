```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('final_submission.txt',delimiter=';',header=None)
df1.columns = ['actual','predict']
```

```python
original = df1.copy()
```

```python
n = len(original)
```

```python
actual = []
predict = []

for i in range(n):
    actual.append(np.array(original)[i][0][1:-1].split(', '))
    predict.append(np.array(original)[i][1][1:-1].split(', '))    
        
```

```python
original['actual']=actual
original['predict']=predict
```

```python
for i in range(n):
    while len(predict[i])>4:
        try:
            predict[i].remove('None')
        except:
            try: 
                predict[i].remove('3')
            except:
                predict[i].pop()
        print(actual[i],predict[i])
```

```python
results = np.array(actual)==np.array(predict)
```

```python
print('original results =', np.sum(results*1)/results.size)
```
